```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

# Output directory for charts
IMAGE_DIR = "../reports/images"
os.makedirs(IMAGE_DIR, exist_ok=True)

# Set styles
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 10
print("Libraries imported and setup complete.")
```

```python
# Load cleaned datasets from D:\mutual-fund-analytics\data\processed
nav_df = pd.read_csv("../data/processed/02_nav_history.csv")
master_df = pd.read_csv("../data/processed/01_fund_master.csv")
aum_df = pd.read_csv("../data/processed/03_aum_by_fund_house.csv")
sip_df = pd.read_csv("../data/processed/04_monthly_sip_inflows.csv")
cat_inflow_df = pd.read_csv("../data/processed/05_category_inflows.csv")
folio_df = pd.read_csv("../data/processed/06_industry_folio_count.csv")
perf_df = pd.read_csv("../data/processed/07_scheme_performance.csv")
tx_df = pd.read_csv("../data/processed/08_investor_transactions.csv")
holding_df = pd.read_csv("../data/processed/09_portfolio_holdings.csv")
bench_df = pd.read_csv("../data/processed/10_benchmark_indices.csv")
print("All 10 datasets loaded successfully.")
```

### Chart 1: Daily NAV Trend Analysis (2022 - 2026)
Plots daily NAV for all 40 schemes, highlighting the 2023 Bull Run and 2024 Market Corrections.

```python
# Merge NAV with Master to get Scheme Names
nav_merged = nav_df.merge(master_df[['amfi_code', 'scheme_name']], on='amfi_code')
fig1 = px.line(nav_merged, x='date', y='nav', color='scheme_name', title='Daily NAV Trend Analysis (2022-2026)')
fig1.add_vrect(
    x0="2023-04-01", x1="2023-12-31",
    fillcolor="green", opacity=0.1, line_width=0,
    annotation_text="2023 Bull Run", annotation_position="top left"
)
fig1.add_vrect(
    x0="2024-01-01", x1="2024-06-30",
    fillcolor="red", opacity=0.1, line_width=0,
    annotation_text="2024 Corrections", annotation_position="top left"
)
fig1.update_layout(showlegend=False, xaxis_title="Date", yaxis_title="NAV Value")
try:
    fig1.write_image("../reports/images/01_nav_trend_plotly.png")
except Exception as e:
    print("Kaleido write_image failed, creating fallback Matplotlib line chart...")
    plt.figure(figsize=(12, 6))
    for name, group in nav_merged.groupby('scheme_name'):
        plt.plot(pd.to_datetime(group['date']), group['nav'], alpha=0.3)
    plt.axvspan(pd.to_datetime('2023-04-01'), pd.to_datetime('2023-12-31'), color='green', alpha=0.1, label='2023 Bull Run')
    plt.axvspan(pd.to_datetime('2024-01-01'), pd.to_datetime('2024-06-30'), color='red', alpha=0.1, label='2024 Corrections')
    plt.title('Daily NAV Trend Analysis (2022-2026)')
    plt.xlabel('Date')
    plt.ylabel('NAV')
    plt.savefig("../reports/images/01_nav_trend_plotly.png", dpi=150)
    plt.close()
fig1.show()
```

### Chart 2: AUM Growth by Fund House (2022 - 2025)
Grouped bar chart showing AUM across fund houses per year, highlighting SBI's peak dominance.

```python
aum_df['year'] = pd.to_datetime(aum_df['date']).dt.year
aum_grouped = aum_df.groupby(['year', 'fund_house'])['aum_lakh_crore'].sum().reset_index()

plt.figure(figsize=(12, 7))
ax = sns.barplot(data=aum_grouped, x='year', y='aum_lakh_crore', hue='fund_house', palette='muted')
plt.title("AUM Growth by Fund House (2022 - 2025)")
plt.ylabel("AUM (Lakh Crore)")
plt.xlabel("Year")

# Annotate SBI Dominance
plt.annotate('SBI Dominance: ~₹6L+ Cr (2022-2025)', 
             xy=(3.15, 6.05), xytext=(1.5, 7.5),
             arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6))

plt.tight_layout()
plt.savefig("../reports/images/02_aum_growth_seaborn.png", dpi=150)
plt.show()
```

### Chart 3: Monthly SIP Inflow Trend (2022 - 2025)
Time-series trace annotating the ₹31,002 Cr record peak inflow in December 2025.

```python
fig3 = px.line(sip_df, x='month', y='sip_inflow_crore', title='Monthly SIP Inflow Trend (2022-2025)', markers=True)
dec_25_row = sip_df[sip_df['month'] == '2025-12']
if not dec_25_row.empty:
    val = dec_25_row['sip_inflow_crore'].values[0]
    fig3.add_annotation(
        x='2025-12', y=val,
        text=f"Record High: ₹{val:,} Cr",
        showarrow=True, arrowhead=1, ax=-40, ay=-40
    )
fig3.update_layout(xaxis_title="Month", yaxis_title="SIP Inflow (Crore)")
try:
    fig3.write_image("../reports/images/03_sip_inflow_plotly.png")
except Exception as e:
    plt.figure(figsize=(10, 6))
    plt.plot(sip_df['month'], sip_df['sip_inflow_crore'], marker='o', color='#2B6CB0')
    plt.title('Monthly SIP Inflow Trend (2022-2025)')
    plt.xlabel('Month')
    plt.ylabel('SIP Inflow (Crore)')
    plt.xticks(rotation=45)
    plt.savefig("../reports/images/03_sip_inflow_plotly.png", dpi=150)
    plt.close()
fig3.show()
```

### Chart 4: Net Category Inflows Heatmap (2024 - 2025)
Heatmap visualization illustrating monthly net inflows across core fund categories.

```python
heatmap_data = cat_inflow_df.pivot(index='category', columns='month', values='net_inflow_crore')
plt.figure(figsize=(14, 6))
sns.heatmap(heatmap_data, cmap='Blues', annot=True, fmt='.1f', cbar_kws={'label': 'Net Inflow (Crore)'})
plt.title("Net Category Inflows Heatmap")
plt.xlabel("Month")
plt.ylabel("Category")
plt.tight_layout()
plt.savefig("../reports/images/04_category_inflow_heatmap.png", dpi=150)
plt.show()
```

### Chart 5: Investor Age Group Distribution
Pie chart representing the participation count of different age groups.

```python
age_counts = tx_df.drop_duplicates(subset=['investor_id'])['age_group'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Pastel1"))
plt.title("Investor Age Group Distribution")
plt.tight_layout()
plt.savefig("../reports/images/05_age_distribution_pie.png", dpi=150)
plt.show()
```

### Chart 6: SIP Transaction Amount by Age Group
Box plot showing the variation and median SIP amount values across age groups.

```python
sip_tx = tx_df[tx_df['transaction_type'] == 'SIP']
plt.figure(figsize=(10, 6))
sns.boxplot(data=sip_tx, x='age_group', y='amount_inr', order=sorted(sip_tx['age_group'].unique()), palette='Set3')
plt.title("SIP Transaction Amount by Age Group")
plt.ylabel("SIP Amount (INR)")
plt.xlabel("Age Group")
plt.tight_layout()
plt.savefig("../reports/images/06_sip_amount_by_age_box.png", dpi=150)
plt.show()
```

### Chart 7: Investor Gender Distribution
Pie chart describing gender split among retail mutual fund investors.

```python
gender_counts = tx_df.drop_duplicates(subset=['investor_id'])['gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Pastel2"))
plt.title("Investor Gender Distribution")
plt.tight_layout()
plt.savefig("../reports/images/07_gender_distribution_pie.png", dpi=150)
plt.show()
```

### Chart 8: Total SIP Investment Amount by State
Horizontal bar chart outlining geographical revenue breakdown across states.

```python
state_sip = sip_tx.groupby('state')['amount_inr'].sum().reset_index().sort_values(by='amount_inr', ascending=False)
plt.figure(figsize=(10, 8))
sns.barplot(data=state_sip, x='amount_inr', y='state', palette='viridis')
plt.title("Total SIP Investment Amount by State")
plt.xlabel("Total SIP Amount (INR)")
plt.ylabel("State")
plt.tight_layout()
plt.savefig("../reports/images/08_sip_amount_by_state_bar.png", dpi=150)
plt.show()
```

### Chart 9: City Tier Distribution (T30 vs B30)
Pie chart representing geographic demographic tiering (Top 30 vs Beyond 30 cities).

```python
t30_b30 = tx_df.drop_duplicates(subset=['investor_id'])['city_tier'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(t30_b30, labels=t30_b30.index, autopct='%1.1f%%', startangle=140, colors=['#4FD1C5', '#FC8181'])
plt.title("City Tier Distribution: T30 vs B30")
plt.tight_layout()
plt.savefig("../reports/images/09_t30_b30_pie.png", dpi=150)
plt.show()
```

### Chart 10: Industry Folio Count Growth (2022 - 2025)
Line chart tracing retail participation growth from 13.26 Cr in Jan 2022 to 26.12 Cr in Dec 2025.

```python
plt.figure(figsize=(10, 6))
plt.plot(folio_df['month'], folio_df['total_folios_crore'], marker='o', linewidth=2, color='#3182CE')
plt.title("Industry Folio Count Growth (2022 - 2025)")
plt.xlabel("Month")
plt.ylabel("Total Folios (Crore)")

start_val = folio_df.iloc[0]['total_folios_crore']
end_val = folio_df.iloc[-1]['total_folios_crore']
plt.annotate(f"Start: {start_val} Cr", xy=(0, start_val), xytext=(2.0, start_val + 1),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5))
plt.annotate(f"End: {end_val} Cr", xy=(len(folio_df)-1, end_val), xytext=(len(folio_df)-4, end_val - 2),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5))
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("../reports/images/10_folio_growth_line.png", dpi=150)
plt.show()
```

### Chart 11: Daily Returns Pairwise Correlation Matrix
Pairwise returns correlation heatmap for 10 selected mutual fund schemes.

```python
selected_codes = master_df['amfi_code'].head(10).tolist()
nav_selected = nav_df[nav_df['amfi_code'].isin(selected_codes)]
nav_pivoted = nav_selected.pivot(index='date', columns='amfi_code', values='nav')
returns_df = nav_pivoted.pct_change().dropna()
code_to_name = master_df.set_index('amfi_code')['scheme_name'].to_dict()
returns_df = returns_df.rename(columns=code_to_name)
corr_matrix = returns_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Pairwise NAV Return Correlation Matrix (10 Selected Funds)")
plt.tight_layout()
plt.savefig("../reports/images/11_return_correlation_heatmap.png", dpi=150)
plt.show()
```

### Chart 12: Aggregate Sector Allocation across Equity Funds
Donut chart demonstrating sector weight distributions.

```python
sector_weights = holding_df.groupby('sector')['weight_pct'].sum().reset_index()
sector_weights = sector_weights.sort_values(by='weight_pct', ascending=False)
top_sectors = sector_weights.head(8)
others_weight = sector_weights.iloc[8:]['weight_pct'].sum()
if others_weight > 0:
    others_df = pd.DataFrame([{'sector': 'Others', 'weight_pct': others_weight}])
    top_sectors = pd.concat([top_sectors, others_df], ignore_index=True)

plt.figure(figsize=(8, 8))
plt.pie(top_sectors['weight_pct'], labels=top_sectors['sector'], autopct='%1.1f%%', 
        startangle=90, colors=sns.color_palette("tab10"), pctdistance=0.85)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Aggregate Sector Allocation across Equity Funds")
plt.tight_layout()
plt.savefig("../reports/images/12_sector_allocation_donut.png", dpi=150)
plt.show()
```

### Charts 13 - 16: Additional Supporting Analytical Visualizations
Additional charts demonstrating transaction values, holding market capitalization, risk grades, and trailing performance spreads.

```python
# Chart 13: Transaction values distribution
plt.figure(figsize=(10, 6))
sns.histplot(tx_df['amount_inr'], bins=50, kde=True, color='#E53E3E')
plt.title("Distribution of Investor Transaction Amounts")
plt.xlabel("Transaction Amount (INR)")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("../reports/images/13_transaction_amount_distribution.png", dpi=150)
plt.show()

# Chart 14: Top 10 Stocks by market value
stock_val = holding_df.groupby('stock_name')['market_value_cr'].sum().reset_index().sort_values(by='market_value_cr', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(data=stock_val, x='market_value_cr', y='stock_name', palette='crest')
plt.title("Top 10 Stock Holdings by Aggregate Market Value (Crores)")
plt.xlabel("Total Market Value (Crores)")
plt.ylabel("Stock Name")
plt.tight_layout()
plt.savefig("../reports/images/14_top_stock_holdings_bar.png", dpi=150)
plt.show()

# Chart 15: Risk Category distribution
risk_counts = master_df['risk_category'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=risk_counts.index, y=risk_counts.values, palette='Oranges_r')
plt.title("Distribution of Funds across Risk Categories")
plt.xlabel("Risk Category")
plt.ylabel("Number of Funds")
plt.tight_layout()
plt.savefig("../reports/images/15_risk_distribution_bar.png", dpi=150)
plt.show()

# Chart 16: Return Spread Boxplot
melted_returns = perf_df.melt(id_vars=['scheme_name'], value_vars=['return_1yr_pct', 'return_3yr_pct', 'return_5yr_pct'],
                              var_name='Period', value_name='Return_Pct')
plt.figure(figsize=(10, 6))
sns.boxplot(data=melted_returns, x='Period', y='Return_Pct', palette='Set2')
plt.title("Distribution of Trailing Returns (1-Year, 3-Year, 5-Year)")
plt.ylabel("Return (%)")
plt.xlabel("Time Period")
plt.tight_layout()
plt.savefig("../reports/images/16_returns_distribution_box.png", dpi=150)
plt.show()
```

## 10 Key Business Insights & Findings

1. **NAV Trajectory Bull Run and Corrections:** Daily NAV trend analysis reveals a strong upward trajectory across all 40 schemes during the 2023 market bull run, followed by structural volatility and corrections in the first half of 2024. *(Reference Chart: 01_nav_trend_plotly.png)*
2. **AUM Asset Dominance:** SBI Mutual Fund maintains clear market share dominance in assets under management (AUM) compared to peer AMCs across all years (2022-2025), peaking at approximately ₹6.05L+ Cr. *(Reference Chart: 02_aum_growth_seaborn.png)*
3. **SIP Inflows Acceleration:** Monthly industry-wide SIP inflow trends show exponential growth, rising from ~₹11,517 Cr in Jan 2022 to an all-time record high of ₹31,002 Cr in December 2025. *(Reference Chart: 03_sip_inflow_plotly.png)*
4. **Capital Net Inflow Concentration:** Category net inflows show consistent positive net capital inflows for Large Cap, Mid Cap, and Small Cap funds, with debt funds experiencing higher cyclical shifts. *(Reference Chart: 04_category_inflow_heatmap.png)*
5. **Younger Demographics Concentration:** Retail mutual fund participation is heavily skewed towards younger age brackets, with the 18–30 and 31–45 groups representing over 75% of the total investor base. *(Reference Chart: 05_age_distribution_pie.png)*
6. **Income and Investment Ticket Uniformity:** SIP monthly ticket sizes are relatively uniform across age groups, though the 31–45 age group has a slightly higher median investment amount. *(Reference Chart: 06_sip_amount_by_age_box.png)*
7. **Investor Gender Disparity:** Retail mutual fund participation is highly dominated by male investors (~74.8%), suggesting that targeted financial inclusion programs for women are highly needed. *(Reference Chart: 07_gender_distribution_pie.png)*
8. **Geographical Concentration:** Punjab, Tamil Nadu, and Madhya Pradesh stand out as the top three geographic contributors by total transaction values. *(Reference Chart: 08_sip_amount_by_state_bar.png)*
9. **City Tier Penetration:** Top 30 (T30) cities represent over 70% of total retail transaction volumes, with Beyond 30 (B30) cities accounting for the remaining 30%, indicating room for rural expansion. *(Reference Chart: 09_t30_b30_pie.png)*
10. **Retail Onboarding Surge:** The industry's folio counts experienced massive expansion, doubling from 13.26 Cr in January 2022 to 26.12 Cr in December 2025, showing strong retail onboarding. *(Reference Chart: 10_folio_growth_line.png)*
